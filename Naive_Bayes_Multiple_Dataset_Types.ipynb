```python
def average(dataset):
    mean = float(sum(dataset)/len(dataset))
    return mean
```

```python
def norm_dist(data, mu, sigma):
    constant = 1/sqrt(2*pi*sigma**2)
    exponent = exp((-(data - mu)**2)/(2*sigma**2))
    norm = constant*exponent
    return norm
```

```python
def naive_bayes(train, test):
    #Set up some variables that will allow us
    #To put in any dataset and only change the
    #Name of the arrays.
    train = np.asarray(train)
    test = np.asarray(test)
    ncol = len(train[0])
    nrows_train = len(train)
    nrows_test = len(test)
    class_test = np.asarray(test[:,ncol-1])
    test = np.delete(test, ncol-1, -1)
    train[:,ncol-1] = train[:,ncol-1].astype(int)
    mini = int(np.amin(train[:,ncol-1]))
    maxi = int(np.amax(train[:,ncol-1]))+1
    #We need to calculate the mean and standard deviation for each column
    #In the different classes. We will use these to get the gaussian 
    #Probabilities. We also need the percentage of classes
    mean = np.array([])
    std = np.array([])
    class_prob = np.array([])
    for class_num in range(mini,maxi):
        mat = np.zeros((1,ncol))
        for row in range(0,nrows_train):
            if train[row,ncol-1] == class_num:
                mat = np.vstack((mat, train[row]))
        mat = np.delete(mat, 0, axis = 0)
        if mat.size == 0:
            for column in range(ncol-1):
                av = 0
                mean = np.append(mean,av)
                std_ensure = 0.01
                std = np.append(std, std_ensure)
        else:
            for column in range(ncol-1):
                av = average(mat[:,column])
                mean = np.append(mean, av)
                std_ensure = np.std(mat[:,column])
                if std_ensure == 0:
                    std_ensure = 0.01
                std = np.append(std, std_ensure)#np.std(mat[:,column]))
        nrows_mat = len(mat)
        class_prob = np.append(class_prob, (nrows_mat/nrows_train))
    #Now that we have calculated the mean and standard deviation
    #For each classification. We will us two for loop to make the 
    #Necessary print statements
    if mini == 1: 
         maxi -= 1
    mean = np.reshape(mean, (maxi, ncol-1))
    std = np.reshape(std, (maxi, ncol-1))

    for i in range(0,maxi):
        for j in range(ncol-1):
            if mini == 1:
                index = i+1
            else:
                index = i
            print("Class", index,", Attribute", j+1,", Mean =", round(mean[i,j],2),", Std =", round(std[i,j],2))
    class_prob = np.reshape(class_prob,(maxi,1))
    prob = np.array([])
    pred = np.array([])
    acc = 0
    for i in range(nrows_test):
        prob_by_class = np.array([])
        classes = np.array([])
        act = class_test[i]
        for z in range(0, maxi):
            norm_prob = 0
            for j in range(ncol-2):
                normal = norm_dist(test[i,j], mean[z,j], std[z,j])
                norm_prob += np.log(normal)
                final = norm_prob + np.log(class_prob[z])
            prob_by_class = np.append(prob_by_class, final)
        for x in range(0,len(prob_by_class)):
            if prob_by_class[x] == np.amax(prob_by_class):
                if mini == 1:
                    x = x+1
                classes = np.append(classes, x)
        if len(classes) > 1:
            length = len(classes)
        marker = np.intersect1d(act,classes) 
        if marker.size > 0:
            acc = acc
            if len(classes) > 1 and marker == act:
                acc += 1/length
            if len(classes) > 1 and marker != act:
                acc = acc
            if len(classes) == 1 and marker == act:
                acc += 1
        else:
            acc = acc
        pred = np.append(pred, np.random.choice(classes))
        prob = np.append(prob, np.amax(prob_by_class))
        print("ID", i+1, ", Predicted =", pred[i], ", Probability =", round(prob[i], 2), 
              ", True =", act, ", Accuracy =", round(acc/nrows_test,2))
    return acc/nrows_test
```

```python
pen_acc = naive_bayes(pen_train, pen_test)
```

```python
sat_acc = naive_bayes(sat_train, sat_test)
```

```python
yeast_acc = naive_bayes(yeast_train, yeast_test)
```

```python
classification_accuracy = ((pen_acc+sat_acc+yeast_acc)/3)*100
print("Classification Accuracy:", round(classification_accuracy, 2),"%")
```
